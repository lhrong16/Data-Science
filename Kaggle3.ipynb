```python
#Let's focus on 'If your job title needs additional context, please clarify here:' column
# We can select a specific column by using the code df[['column_name']]. Example below
x = df[['If your job title needs additional context, please clarify here:']]
x.head()
#The result takes only one column as below. Lots of NaN/Null values.
```

```python
# If we want to select a subset of 3 columns we can do df[['column_1','column_2','column_3']]
x = df[['How old are you?','Job title','What country do you work in?']]
x.head()
```

```python
# But let's say we want to select a row that meets a certain condition.
# Let's say we want to select those whose salary = 55,000
# We use df[df['column_name']==condition value]
x = df[df["What is your annual salary? (You'll indicate the currency in a later question. If you are part-time or hourly, please enter an annualized equivalent -- what you would earn if you worked the job 40 hours a week, 52 weeks a year.)"]=='55,000']
# Here we must be careful, because the datatype for column Salary is 'object' so what ever values in the column can be considered as a string.
# So that is why we wrapped up 55,000 with '' in the command above
x.head()
# Say we want to select those whose salary is below 55,000
# x = df[df['Salary'] <'55,000'] ----we cannot do this because 55,000 is a string, not a number, so cannot do comparison
```

```python
#Convert string column to int. This is a little bit advance. Please ignore the 'warnings' etc
df["What is your annual salary? (You'll indicate the currency in a later question. If you are part-time or hourly, please enter an annualized equivalent -- what you would earn if you worked the job 40 hours a week, 52 weeks a year.)"] = df["What is your annual salary? (You'll indicate the currency in a later question. If you are part-time or hourly, please enter an annualized equivalent -- what you would earn if you worked the job 40 hours a week, 52 weeks a year.)"].replace(',','',regex=True) #this replace the comma ',' in 55,000 to 55000
df["What is your annual salary? (You'll indicate the currency in a later question. If you are part-time or hourly, please enter an annualized equivalent -- what you would earn if you worked the job 40 hours a week, 52 weeks a year.)"] = df["What is your annual salary? (You'll indicate the currency in a later question. If you are part-time or hourly, please enter an annualized equivalent -- what you would earn if you worked the job 40 hours a week, 52 weeks a year.)"].astype(int) # this convert the column datatypes from object into numbers , in this case integer
df.info()
df.head()
```

```python
#Now only can we select those with salary less than 50000
x = df[df["What is your annual salary? (You'll indicate the currency in a later question. If you are part-time or hourly, please enter an annualized equivalent -- what you would earn if you worked the job 40 hours a week, 52 weeks a year.)"] < 50000] #no need quotation marks anymore, because Salary column is a column of number
x.head()
```

```python
# Back to the question, which column, which row has null values?
# We know the column 'If your job title needs additional context, please clarify here:' has NaN values
x = df['If your job title needs additional context, please clarify here:'] # we assign the column 'If your job title needs additional context, please clarify here:' from the dataframe df to variable
x.info()
```

```python
#To get the rows of the null values
# https://datatofish.com/rows-with-nan-pandas-dataframe/
# When we do x.isna(), it returns a column where if the row is NaN, that row will be marked as True. Else, it will mark the row as False
# When we do x[x.isna()], it makes a list of all rows that is True i.e the NaN rows.
y = x[x.isna()]
y
#As you can see, the length of y is 20,749. This Tallies with our answer above.
```
